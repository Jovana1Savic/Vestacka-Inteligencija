```python
p = [[i + 3*v + 1 for i in range(3)] for v in range(9)]
# Uzima bar jednu vrednost iz domena.
for v in range(9):
    print(p[v][0], p[v][1], p[v][2], 0)
    
# Ne moze imati dve vrednosti istovremeno.
for v in range(9):
    for i in range(3):
        for j in range(i + 1,3):
            if i != j:
                print(-p[v][i], -p[v][j], 0)
                
# Ne sme se ponavljati u istom redu i istoj koloni.
for v in range(9):
    for w in range(9):
        if w != v and (w % 3 == v % 3 or w // 3 == v // 3):
            print(-p[v][0], p[w][1], p[w][2], 0)
            print(-p[v][1], p[w][0], p[w][2], 0)
            print(-p[v][2], p[w][0], p[w][1], 0)
            
# Uslov 6x
B1 = [0, 1, 4]
for v in B1:
    for w in B1:
        if w != v:
            print(-p[v][0], p[w][1], p[w][2], 0)
            print(-p[v][1], p[w][0], p[w][2], 0)
            print(-p[v][2], p[w][0], p[w][1], 0)
            
# Uslov /2
print(p[3][0], p[3][1], 0)
print(p[3][1], p[6][1], 0)
print(p[3][0], p[6][0], 0)
print(p[6][0], p[6][1], 0)

# Uslov /3
print(p[2][0], p[2][2], 0)
print(p[2][2], p[5][2], 0)
print(p[2][0], p[5][0], 0)
print(p[5][0], p[5][2], 0)

# Uslov -1
print(-p[7][1], p[8][0], p[8][2], 0)
print(-p[8][1], p[7][0], p[7][2], 0)
print(p[7][1], p[8][1], 0)
```
